## Exp #1: Using Both Solvers and Propagated IMU
In this expriement, perturb the GT poses to extiate a noisy IMU propagation. At each iteration, the IK solver takes in EKF's propagated pose (perturbed GT) and estimates the f_v and f_h components of the cable forces and an optimized orientation. The optimized orientation and cable forces from the IK solver is then fed into the FK solver alongside the actual cable length measurements to yeild the final FK pose. This pose is expected to match the clean GT pose. 

```python
tr_errors = []
orient_errors = []
fk_poses = []
ik_lcs = []
start_time = time.time()
for i in range(gt_poses.shape[0]):
    #Extract GT poses
    pos_i = gt_poses[i,0:3]
    q_i = gt_poses[i,3:]
    R_i = SO3.from_quaternion(q_i).as_matrix()
    #Set the pulley locations
    robot_params.setPulleyPoses(p1[i,:],p2[i,:],p3[i,:],p4[i,:])
    
    # perturb the GT poses to simulate a propagated IMU
    R_rand = SO3.exp(np.random.randn(3)*0.5*np.pi/180).as_matrix()
    R_fk_init = R_i @ R @ R_rand
    R_imu = R @ R_rand
    r_rand = np.random.randn(3)*0.1
    T_imu = np.array(pos_i) + r_rand
    # Use the IMU pose to get the cable forces and optimized R
    R_ik, lc_cat, f_ik, C1, C2, b_in_w \
    = scampi_ks_solver.inverseKinematicsSolver(robot_params, T_imu, R_imu)
    #Extract offset corrected cable lengths 
    lc_meas = data['measured_cable_lengths'][i,:]
    # Use the optimized R_ik, propagated IMU position, cable forces f_ik, and real cable length measurements to run the second solver
    fk_R, fk_r, fk_fc, fk_C1, fk_C2, fk_b_in_w\
    = scampi_ks_solver.forwardKinematicsSolver(robot_params, lc_meas, f_ik[:,0], T_imu, R_ik)
        
    fk_poses.append(fk_r)
    #construct the error as the difference between the clean GT and estimated pose by the FK
    e_r = (fk_r.squeeze() - pos_i).reshape(3,1)
    translation_error = e_r.T
    orientation_error = np.linalg.norm(SO3(fk_R.T@R).log())*180/np.pi  
    tr_errors.append(translation_error)
    orient_errors.append(orientation_error)

end_time = time.time()
tr_errors = np.vstack(tr_errors)
orient_errors = np.vstack(orient_errors)
fk_poses = np.hstack(fk_poses).T
print(f'process time for each pair of forward and inverse solvers is on average {(end_time-start_time)} ms')
```

## Exp #2: Standalone Operation Using Just the FK Solver
In this expriement, only the cable length measurements are used to solve the forward kinematics. In the first iteration, the known intial pose is given to the IK sover to estimate the initial cable forces for the FK solver. From now on, the IK sover will not be used. Then, at each iteration, the FK solver takes its own previous pose and cable force estimates alongside the actual cable length measurements to yeild the next FK pose. This FK pose is expected to match the clean GT pose. 

```python
tr_errors = []
orient_errors = []
fk_poses = []
ik_lcs = []
start_time = time.time()
fk_fc = None
for i in range(gt_poses.shape[0]):
    pos_i = gt_poses[i,0:3]
    q_i = gt_poses[i,3:]
    R_i = SO3.from_quaternion(q_i).as_matrix()
    
    robot_params.setPulleyPoses(p1[i,:],p2[i,:],p3[i,:],p4[i,:])
    lc_meas = data['measured_cable_lengths'][i,:]

    if fk_fc is None:
        R, lc_cat, ik_fc, C1, C2, b_in_w \
        = scampi_ks_solver.inverseKinematicsSolver(robot_params, pos_i, R_i)
        previous_pos = pos_i
        previous_rot = R_i
        fk_fc = ik_fc

    fk_R, fk_r, fk_fc, fk_C1, fk_C2, fk_b_in_w\
    = scampi_ks_solver.forwardKinematicsSolver(robot_params, lc_meas, fk_fc[:,0], previous_pos, previous_rot)
        
    fk_poses.append(fk_r)
    e_r = (fk_r.squeeze() - pos_i).reshape(3,1)
    translation_error = e_r.T
    orientation_error = np.linalg.norm(SO3(fk_R.T@R).log())*180/np.pi  
    tr_errors.append(translation_error)
    orient_errors.append(orientation_error)

end_time = time.time()
tr_errors = np.vstack(tr_errors)
orient_errors = np.vstack(orient_errors)
fk_poses = np.hstack(fk_poses).T
print(f'process time for each pair of forward and inverse solvers is on average {(end_time-start_time)} ms')
```
